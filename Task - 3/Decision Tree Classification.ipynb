## Decision Tree classification

### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Loading the Iris dataset

```python
dataset=pd.read_csv(r'D:/SHRUT/study/DATA SET/Iris.csv',index_col=0)
```

```python
dataset
```

```python
dataset.info()
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
X
```

```python
y.reshape(len(y),1)
```

### Encoding the dependent variable

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

### Splitting dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier(random_state=0)
dc.fit(X_train,y_train)
```

```python
y_pred=dc.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Checking the accuracy_score,classification_report and confusion_matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

### Visualising the decision tree

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(25,25))

tree_image = plot_tree(dc,feature_names=dataset.columns,class_names=dataset['Species'].unique().tolist(),
                       precision=4,label="all",filled=True,rounded=True)

plt.show()
```
